```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# this is the URL we want to apply scraping
url = "https://www.canada.ca/en/public-health/services/surveillance/respiratory-virus-detections-canada/2018-2019/respiratory-virus-detections-isolations-week-01-ending-january-5-2019.htm"

# Learned that i can use Pandas to read HTML tables directly and a built in function to look for table tags

# scraping the tables
tables = pd.read_html(url)
# i put the table 2 in flu_data
flu_data = tables[1]
    
    
#flu_data.shape
flu_data.info()
print(flu_data.head())
print(flu_data.tail(15))




```

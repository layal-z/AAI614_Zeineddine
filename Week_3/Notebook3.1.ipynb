## Random Walk

```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import datetime
import random
import re

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
random.seed(datetime.datetime.now().strftime('%s'))
def getLinks(articleUrl):
    req = Request(f'http://en.wikipedia.org{articleUrl}', headers=headers)
    html = urlopen(req)
    bs = BeautifulSoup(html, 'html.parser')
    return bs.find('div', {'id':'bodyContent'}).find_all('a', href=re.compile('^(/wiki/)((?!:).)*$'))

links = getLinks('/wiki/Kevin_Bacon')
while len(links) > 0:
    newArticle = links[random.randint(0, len(links)-1)].attrs['href']
    print(newArticle)
    links = getLinks(newArticle)
```

## Recursively crawling an entire site

```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import re

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
pages = set()
def getLinks(pageUrl):
  # Stop if we have gathered 50 unique pages
    if len(pages) >= 50:
        print("Reached data collection limit. Stopping...")
        return


    req = Request(f'http://en.wikipedia.org{pageUrl}', headers=headers)
    html = urlopen(req)
    bs = BeautifulSoup(html, 'html.parser')
    for link in bs.find_all('a', href=re.compile('^(/wiki/)')):
        if 'href' in link.attrs:
            if link.attrs['href'] not in pages:
                #We have encountered a new page
                newPage = link.attrs['href']
                print(newPage)
                pages.add(newPage)
                getLinks(newPage)
getLinks('')
```

## Collecting Data Across an Entire Site

```python
from urllib.error import HTTPError
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import re
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
pages = set()
def getLinks(pageUrl):
  # Because it runs forever, addig STOP CONDITION: Kill the crawl once we have 50 unique pages
  if len(pages) >= 50:
    return
    #because the code returned error since there was a missing http i added a nested try except to account for missing http
  try:
    req=Request(f'http://en.wikipedia.org{pageUrl}', headers=headers)
    html = urlopen(req)
    bs = BeautifulSoup(html, 'html.parser')
    try:
      print(bs.h1.get_text())
      #mw-parser-output
      bodyContent = bs.find('div', {'id':'bodyContent'}).find_all('p')
      if len(bodyContent):
        print(bodyContent[0])
        print(bs.find(id='ca-edit').find('a').attrs['href'])
    except AttributeError:
        print('This page is missing something! Continuing.')
  #for a cleaner crawl add ((?!:).)*$ to filter out specials
    for link in bs.find_all('a', href=re.compile('^(/wiki/)((?!:).)*$')):
      if 'href' in link.attrs:
        if link.attrs['href'] not in pages:
          #We have encountered a new page
          newPage = link.attrs['href']
          print('-'*20)
          print(newPage)
          pages.add(newPage)
          getLinks(newPage)

  except HTTPError as e:
        print(f"Data source missing (404/403): {e}")


getLinks('/wiki/General-purpose_programming_language')
```

## Crawling across the Internet

```python
import time # this use to create a delay
from urllib.request import urlopen, Request
from urllib.parse import urlparse
from bs4 import BeautifulSoup
import re
import datetime
import random
ssl._create_default_https_context = ssl._create_unverified_context
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
pages= set()
Max_PAges = 10
#Retrieves a list of all Internal links found on a page
def getInternalLinks(bs, url):
    netloc = urlparse(url).netloc
    scheme = urlparse(url).scheme
    internalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc == '':
            internalLinks.add(f'{scheme}://{netloc}/{link.attrs["href"].strip("/")}')
        elif parsed.netloc == netloc:
            internalLinks.add(link.attrs['href'])
    return list(internalLinks)

#Retrieves a list of all external links found on a page
def getExternalLinks(bs, url):
    netloc = urlparse(url).netloc
    externalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        if parsed.netloc != '' and parsed.netloc != netloc:
            externalLinks.add(link.attrs['href'])
    return list(externalLinks)

def getRandomExternalLink(startingPage):
  try:
    req = Request(startingPage, headers=headers)
    #adding slow down to avoid 999 error
    time.sleep(1)
    bs = BeautifulSoup(urlopen(req), 'html.parser')
    externalLinks = getExternalLinks(bs, startingPage)

    if not len(externalLinks):
        print('No external links, looking around the site for one')
        internalLinks = getInternalLinks(bs, startingPage)
        #adding this to avoid empty links
        if not internalLinks:
          print("Dead end reached! No links found at all.")
          return None


        return getRandomExternalLink(random.choice(internalLinks))
    else:
        return random.choice(externalLinks)
  except Exception as e:
    print(f"Error accessing {startingPage}: {e}")
    return None


def followExternalOnly(startingSite):
  # to VOID ENDLIESS RUNNNg
  if len(pages)>=Max_PAges:
    print(f"\n--- Limit reached ({Max_PAges} pages). Stopping. ---")
    return

  externalLink = getRandomExternalLink(startingSite)\
#print(f'Random external link is: {externalLink}')
#followExternalOnly(externalLink)

  if externalLink:
    if externalLink not in pages:
      pages.add(externalLink)
      print(f'Random external link is: {externalLink}')
      followExternalOnly(externalLink)
    else:
      print("Duplicate site found, skipping to avoid loops.")
      # Try to find a different link from the current site
      followExternalOnly(startingSite)
  else:
    print("Crawl ended due to a dead end or error.")


followExternalOnly('https://www.oreilly.com/')
```

## Collect all External Links from a Site

```python
# Collects a list of all external URLs found on the site
allExtLinks = []
allIntLinks = []


def getAllExternalLinks(url):
    bs = BeautifulSoup(urlopen(url), 'html.parser')
    internalLinks = getInternalLinks(bs, url)
    externalLinks = getExternalLinks(bs, url)
    for link in externalLinks:
        if link not in allExtLinks:
            allExtLinks.append(link)
            print(link)

    for link in internalLinks:
        if link not in allIntLinks:
            allIntLinks.append(link)
            getAllExternalLinks(link)


allIntLinks.append('https://oreilly.com')
getAllExternalLinks('https://www.oreilly.com/')
```

This code crawls a website by following internal links and collecting external ones it finds along the way, while avoiding duplicates. It works, but it has no stopping rule or protection against blocks, so it can easily run for too long or fail on restricted sites. Im aware of these limitations, but I didnt fix them here since the goal was to understand the crawling logic itself.
